```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import the CSV File
df_beers = pd.read_csv("beers.csv")
```

```python
#Examine the Beer CSV Data
df_beers.head()
```

```python
#Pull Data of Most 
top_beers = df_beers['style'].value_counts().rename_axis('Beers').reset_index(name='Total')
top_beers.head(10)
```

```python
top_10_beers = top_beers.head(10)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
labels = top_10_beers['Beers']
totals = top_10_beers['Total']
ax.pie(totals, labels = labels, autopct='%1.2f%%')
plt.title('Top 10 Beers Consumed by Americans')
plt.savefig('Top10.png', bbox_inches='tight')
plt.show()
```
