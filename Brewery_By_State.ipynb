```python
 # Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File Path
csv = "brewery_count_by_state.csv"

# Import the CSV file as a DataFrame
brewery_count = pd.read_csv(csv, encoding="utf-8")
brewery_count
```

```python
# Clean the Data
brewery_count=brewery_count.drop([0])
brewery_count
```

```python
brewery_count=brewery_count.drop([53])
brewery_count
```

```python
brewery_count=brewery_count.drop([52])
brewery_count
```

```python
brewery_count=brewery_count.drop([54])
brewery_count
```

```python
brewery_count["2022"]=brewery_count["2022"].astype(str).str.replace(",","").astype(int)
brewery_count["2021"]=brewery_count["2021"].astype(str).str.replace(",","").astype(int)
brewery_count["2020"]=brewery_count["2020"].astype(str).str.replace(",","").astype(int)
brewery_count["2019"]=brewery_count["2019"].astype(str).str.replace(",","").astype(int)
brewery_count["2018"]=brewery_count["2018"].astype(str).str.replace(",","").astype(int)
brewery_count["2017"]=brewery_count["2017"].astype(str).str.replace(",","").astype(int)
brewery_count["2016"]=brewery_count["2016"].astype(str).str.replace(",","").astype(int)
brewery_count["2015"]=brewery_count["2015"].astype(str).str.replace(",","").astype(int)
brewery_count["2014"]=brewery_count["2014"].astype(str).str.replace(",","").astype(int)
brewery_count["2013"]=brewery_count["2013"].astype(str).str.replace(",","").astype(int)
brewery_count["2012"]=brewery_count["2012"].astype(str).str.replace(",","").astype(int)
brewery_count["2011"]=brewery_count["2011"].astype(str).str.replace("*","0").astype(int)
brewery_count["2010"]=brewery_count["2010"].astype(str).str.replace(",","").astype(int)
brewery_count
```

```python
brewery_count = brewery_count.sort_values(by = "2022", ascending=False)
brewery_count.head(5).plot(x="STATE",y="2022",kind="bar")
```
